1) - Scatter Plot  


```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

Observations :
               From the scatter plot, it is clear that different species form distinct clusters. Petal length is a strong feature for separating species.

2) - Histrogram
    Observations:

```python
plt.figure(figsize=(8,5))
df['sepal_length'].hist()
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

Observation :
              The sepal length is approximately normally distributed with slight variation.

3)- Boxplot

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Sepal Length by Species")
plt.show()
```

Observation :
              There are a few minor outliers, but overall data spread           is consistent across species.
